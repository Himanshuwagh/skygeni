# SkyGeni Sales Intelligence Analysis
## Part 1: Problem Framing & Exploratory Data Analysis

---

### Business Context
A B2B SaaS company's CRO has reported:
> *"Our win rate has dropped over the last two quarters, but pipeline volume looks healthy. I don't know what exactly is going wrong or what my team should focus on."*

Our goal is to investigate this problem and design a data-driven insight system that helps the CRO understand what's happening and what actions to take.

---

## 🎯 Part 1: Problem Framing (No Code Required)

Before diving into analysis, let's frame the business problem properly. This is critical for designing an effective AI-driven insight system.

### 1.1 What is the REAL Business Problem Here?

**Surface-Level Problem:** Win rate is declining despite healthy pipeline volume.

**Deeper Business Problems to Explore:**

1. **Conversion Efficiency Gap**
   - The sales team is generating leads but failing to convert them effectively
   - This suggests potential issues in the *middle and late stages* of the sales funnel
   - Money and effort are being invested in pipeline generation, but the ROI is decreasing

2. **Revenue Predictability Crisis**
   - A declining win rate makes revenue forecasting unreliable
   - This affects capacity planning, hiring decisions, and investor communications
   - The CRO needs to understand *why* to provide accurate guidance to the board

3. **Resource Misallocation**
   - Sales reps might be spending time on low-probability deals
   - High-potential deals might not be getting adequate attention
   - Without knowing the root cause, optimization is impossible

4. **Potential Market/Competitive Shifts**
   - The decline could signal external factors (new competitors, market saturation)
   - Or internal factors (product-market fit issues in certain segments)
   - Understanding the "where" is as important as the "what"

**The REAL Problem:** The CRO lacks *diagnostic visibility* into what's driving the win rate decline. They need:
- **Attribution**: Which specific factors (reps, segments, deal types) are responsible?
- **Actionability**: What can be changed to reverse the trend?
- **Prioritization**: Where should limited resources be focused for maximum impact?

### 1.2 What Key Questions Should an AI System Answer for the CRO?

An effective sales intelligence system should answer questions across **three dimensions**:

---

#### **A. DIAGNOSTIC Questions (What's Happening & Why?)**

| Question | Business Value |
|----------|---------------|
| Where exactly is the win rate dropping? (By industry, region, product type, rep) | Targeted intervention |
| At which stage of the sales funnel are deals being lost? | Process optimization |
| Has the quality of leads changed? (Lead source performance) | Marketing alignment |
| Are certain deal sizes more likely to be lost now vs. before? | Pricing/packaging review |
| How has the competitive landscape changed? (Win/loss reasons if available) | Strategic response |
| Are sales cycles getting longer? Does this correlate with losses? | Deal health signals |

---

#### **B. PREDICTIVE Questions (What Will Happen?)**

| Question | Business Value |
|----------|---------------|
| Which current pipeline deals are at high risk of being lost? | Early intervention |
| What's the realistic forecast for next quarter given current trends? | Accurate planning |
| Which segments should we prioritize to maximize win rate recovery? | Resource allocation |
| What deal characteristics predict success vs. failure? | Lead scoring |

---

#### **C. PRESCRIPTIVE Questions (What Should We Do?)**

| Question | Business Value |
|----------|---------------|
| Which specific reps need coaching, and on what? | Performance management |
| Should we adjust our ICP (Ideal Customer Profile) focus? | Go-to-market strategy |
| What nurturing strategies work for at-risk deals? | Playbook optimization |
| Are there quick wins - segments where small changes yield big improvements? | ROI maximization |

### 1.3 What Metrics Matter Most for Diagnosing Win Rate Issues?

Let's organize metrics into **tiers** based on diagnostic importance:

---

#### **Tier 1: Primary Diagnostic Metrics** (Must Have)

| Metric | Formula | Why It Matters |
|--------|---------|----------------|
| **Overall Win Rate** | Won Deals / (Won + Lost Deals) | The north star metric |
| **Win Rate by Segment** | Win rate sliced by industry, region, product, rep | Identifies problem areas |
| **Win Rate Trend (QoQ)** | Compare win rates across quarters | Confirms the decline pattern |
| **Stage Conversion Rates** | % of deals progressing from each stage | Funnel leak detection |
| **Average Sales Cycle (ASC)** | Mean days from created to closed | Health indicator |

---

#### **Tier 2: Contributing Factor Metrics** (Important)

| Metric | Formula | Why It Matters |
|--------|---------|----------------|
| **Average Deal Value (Won vs. Lost)** | Mean ACV for each outcome | Deal size correlation |
| **Lead Source Win Rate** | Win rate by lead source (Referral, Inbound, Outbound, Partner) | Lead quality signal |
| **Rep Performance Distribution** | Win rate variance across reps | Coaching needs |
| **Pipeline Velocity** | Deals × Win Rate × Avg Deal Value / Avg Sales Cycle | Overall health |
| **Time-to-Stage** | Days spent in each stage | Stuck deal identification |

---

#### **Tier 3: Supporting Metrics** (Good to Have)

| Metric | Formula | Why It Matters |
|--------|---------|----------------|
| **Volume by Segment** | Deal count by dimension | Mix shift detection |
| **Deal Value Distribution** | Distribution of ACV | Pricing anomalies |
| **Rep Workload** | Active deals per rep | Capacity issues |
| **Win Rate by Deal Stage** | Win rate at entry stage | Deal quality by funnel position |

### 1.4 What Assumptions Are We Making About the Data or Business?

#### **Data Assumptions:**

| Assumption | Risk if False | Mitigation |
|------------|--------------|------------|
| `outcome` field is accurate and final | Misclassified deals skew analysis | Validate with CRM status |
| All deals in dataset represent real opportunities | "Fake" deals inflate pipeline | Check deal value distributions |
| `closed_date` represents actual close, not data entry date | Timing analysis would be wrong | Cross-check with activity logs |
| Deal stages are consistently used across reps | Stage comparison would be invalid | Audit stage usage patterns |
| Win/Lost are the only outcomes (no "abandoned") | Incomplete funnel picture | Check for null/missing outcomes |
| Data covers a representative time period | Conclusions may not generalize | Validate with historical data |

---

#### **Business Context Assumptions:**

| Assumption | Risk if False | Mitigation |
|------------|--------------|------------|
| No major product changes in the analysis period | External factor confounds | Timeline event mapping |
| Pricing remained relatively stable | Price changes affect win rates | Check for pricing tier data |
| Sales process is similar across regions | Regional differences might be normal | Segment benchmarks |
| Competition hasn't dramatically shifted | Market dynamics matter | Competitor intelligence data |
| CRO's "healthy pipeline" claim is accurate | Volume might actually be an issue | Verify pipeline metrics |
| Rep assignments are consistent (no mid-deal handoffs) | Attribution would be wrong | Check for ownership changes |

---

#### **Analytical Assumptions:**

| Assumption | Risk if False | Mitigation |
|------------|--------------|------------|
| Past patterns predict future behavior | Market changes invalidate models | Regular model retraining |
| Win rate is the right metric (not revenue or margin) | Optimizing wrong objective | Align with CRO on goals |
| Correlation implies causation for key factors | Intervention might not work | A/B testing recommendations |
| Sample size per segment is sufficient | Small sample noise | Statistical significance testing |

## 📊 Part 2: Exploratory Data Analysis

Now let's explore the data to validate our hypotheses and discover insights.

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)
plt.style.use('seaborn-v0_8-whitegrid')

# Load the data
df = pd.read_csv('skygeni_sales_data.csv')

print(f"Dataset Shape: {df.shape[0]:,} deals, {df.shape[1]} features")
print(f"\nDate Range: {df['created_date'].min()} to {df['created_date'].max()}")
df.head()
```

```python
# Data overview and quality check
print("=" * 60)
print("DATA QUALITY OVERVIEW")
print("=" * 60)

print("\n📋 Data Types:")
print(df.dtypes)

print("\n\n📋 Missing Values:")
print(df.isnull().sum())

print("\n\n📋 Statistical Summary:")
df.describe()
```

```python
# Categorical field overview
print("=" * 60)
print("CATEGORICAL VARIABLES OVERVIEW")
print("=" * 60)

categorical_cols = ['industry', 'region', 'product_type', 'lead_source', 'deal_stage', 'outcome']

for col in categorical_cols:
    print(f"\n📌 {col.upper()}:")
    print(df[col].value_counts())
    print(f"   Unique values: {df[col].nunique()}")
```

```python
# Convert dates and create time-based features
df['created_date'] = pd.to_datetime(df['created_date'])
df['closed_date'] = pd.to_datetime(df['closed_date'])

# Extract time features
df['created_quarter'] = df['created_date'].dt.to_period('Q')
df['closed_quarter'] = df['closed_date'].dt.to_period('Q')
df['created_month'] = df['created_date'].dt.to_period('M')
df['closed_month'] = df['closed_date'].dt.to_period('M')
df['created_year'] = df['created_date'].dt.year

# Create binary outcome
df['is_won'] = (df['outcome'] == 'Won').astype(int)

print("✅ Date features created")
print(f"\nQuarters covered: {df['closed_quarter'].unique()}")
```

### 2.1 Overall Win Rate Analysis

```python
# Overall Win Rate
overall_win_rate = df['is_won'].mean() * 100
total_won = df['is_won'].sum()
total_lost = len(df) - total_won

print("=" * 60)
print("OVERALL WIN RATE ANALYSIS")
print("=" * 60)
print(f"\n📊 Overall Win Rate: {overall_win_rate:.1f}%")
print(f"   Total Deals Won: {total_won:,}")
print(f"   Total Deals Lost: {total_lost:,}")
print(f"   Total Deals: {len(df):,}")
```

```python
# Win Rate by Quarter (CLOSED date - when deals resolved)
quarterly_metrics = df.groupby('closed_quarter').agg(
    total_deals=('deal_id', 'count'),
    won_deals=('is_won', 'sum'),
    win_rate=('is_won', 'mean'),
    avg_deal_value=('deal_amount', 'mean'),
    total_revenue=('deal_amount', lambda x: x[df.loc[x.index, 'is_won'] == 1].sum()),
    avg_sales_cycle=('sales_cycle_days', 'mean')
).reset_index()

quarterly_metrics['win_rate_pct'] = quarterly_metrics['win_rate'] * 100

print("=" * 60)
print("QUARTERLY WIN RATE TREND (by Closed Date)")
print("=" * 60)
print(quarterly_metrics[['closed_quarter', 'total_deals', 'won_deals', 'win_rate_pct', 'avg_sales_cycle']].to_string(index=False))

# Plotting
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Win Rate Trend
ax1 = axes[0]
quarters = [str(q) for q in quarterly_metrics['closed_quarter']]
ax1.plot(quarters, quarterly_metrics['win_rate_pct'], marker='o', linewidth=2, color='#2E86AB')
ax1.axhline(y=overall_win_rate, color='red', linestyle='--', alpha=0.7, label='Average')
ax1.fill_between(quarters, quarterly_metrics['win_rate_pct'], alpha=0.3)
ax1.set_xlabel('Quarter (Closed Date)')
ax1.set_ylabel('Win Rate (%)')
ax1.set_title('📉 Win Rate Trend by Quarter', fontsize=14, fontweight='bold')
ax1.legend()
ax1.tick_params(axis='x', rotation=45)

# Deal Volume
ax2 = axes[1]
colors = ['#28A745' if w >= overall_win_rate else '#DC3545' for w in quarterly_metrics['win_rate_pct']]
bars = ax2.bar(quarters, quarterly_metrics['total_deals'], color=colors, alpha=0.7)
ax2.set_xlabel('Quarter (Closed Date)')
ax2.set_ylabel('Total Deals Closed')
ax2.set_title('📊 Deal Volume by Quarter', fontsize=14, fontweight='bold')
ax2.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

### 2.2 Segmented Win Rate Analysis

Let's identify WHERE the win rate is dropping.

```python
def calculate_segment_win_rates(df, segment_col, min_deals=30):
    """Calculate win rates by segment with statistical context"""
    result = df.groupby(segment_col).agg(
        total_deals=('deal_id', 'count'),
        won_deals=('is_won', 'sum'),
        win_rate=('is_won', 'mean'),
        avg_deal_value=('deal_amount', 'mean'),
        avg_sales_cycle=('sales_cycle_days', 'mean')
    ).reset_index()
    
    result['win_rate_pct'] = result['win_rate'] * 100
    result = result.sort_values('win_rate_pct', ascending=False)
    result['vs_avg'] = result['win_rate_pct'] - overall_win_rate
    return result

# Win Rate by Industry
print("=" * 60)
print("WIN RATE BY INDUSTRY")
print("=" * 60)
industry_wr = calculate_segment_win_rates(df, 'industry')
print(industry_wr[['industry', 'total_deals', 'won_deals', 'win_rate_pct', 'vs_avg']].to_string(index=False))
```

```python
# Win Rate by Region
print("=" * 60)
print("WIN RATE BY REGION")
print("=" * 60)
region_wr = calculate_segment_win_rates(df, 'region')
print(region_wr[['region', 'total_deals', 'won_deals', 'win_rate_pct', 'vs_avg']].to_string(index=False))
```

```python
# Win Rate by Product Type
print("=" * 60)
print("WIN RATE BY PRODUCT TYPE")
print("=" * 60)
product_wr = calculate_segment_win_rates(df, 'product_type')
print(product_wr[['product_type', 'total_deals', 'won_deals', 'win_rate_pct', 'vs_avg', 'avg_deal_value']].to_string(index=False))
```

```python
# Win Rate by Lead Source
print("=" * 60)
print("WIN RATE BY LEAD SOURCE")
print("=" * 60)
lead_source_wr = calculate_segment_win_rates(df, 'lead_source')
print(lead_source_wr[['lead_source', 'total_deals', 'won_deals', 'win_rate_pct', 'vs_avg']].to_string(index=False))
```

```python
# Visualize segment comparison
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

def plot_segment_bars(ax, data, col, title):
    colors = ['#28A745' if v >= overall_win_rate else '#DC3545' for v in data['win_rate_pct']]
    bars = ax.barh(data[col], data['win_rate_pct'], color=colors, alpha=0.7)
    ax.axvline(x=overall_win_rate, color='navy', linestyle='--', linewidth=2, label=f'Avg: {overall_win_rate:.1f}%')
    ax.set_xlabel('Win Rate (%)')
    ax.set_title(title, fontsize=12, fontweight='bold')
    ax.legend()
    
    # Add value labels
    for bar, val in zip(bars, data['win_rate_pct']):
        ax.text(val + 0.5, bar.get_y() + bar.get_height()/2, f'{val:.1f}%', va='center', fontsize=9)

plot_segment_bars(axes[0,0], industry_wr, 'industry', '📊 Win Rate by Industry')
plot_segment_bars(axes[0,1], region_wr, 'region', '📊 Win Rate by Region')
plot_segment_bars(axes[1,0], product_wr, 'product_type', '📊 Win Rate by Product Type')
plot_segment_bars(axes[1,1], lead_source_wr, 'lead_source', '📊 Win Rate by Lead Source')

plt.tight_layout()
plt.show()
```

### 2.3 Sales Rep Performance Analysis

```python
# Sales Rep Performance
rep_performance = df.groupby('sales_rep_id').agg(
    total_deals=('deal_id', 'count'),
    won_deals=('is_won', 'sum'),
    win_rate=('is_won', 'mean'),
    total_revenue=('deal_amount', lambda x: x[df.loc[x.index, 'is_won'] == 1].sum()),
    avg_deal_value=('deal_amount', 'mean'),
    avg_sales_cycle=('sales_cycle_days', 'mean')
).reset_index()

rep_performance['win_rate_pct'] = rep_performance['win_rate'] * 100
rep_performance = rep_performance.sort_values('win_rate_pct', ascending=False)

print("=" * 60)
print("SALES REP PERFORMANCE SUMMARY")
print("=" * 60)
print(f"\n📊 Rep Performance Distribution:")
print(f"   Highest Win Rate: {rep_performance['win_rate_pct'].max():.1f}% ({rep_performance.iloc[0]['sales_rep_id']})")
print(f"   Lowest Win Rate: {rep_performance['win_rate_pct'].min():.1f}% ({rep_performance.iloc[-1]['sales_rep_id']})")
print(f"   Median Win Rate: {rep_performance['win_rate_pct'].median():.1f}%")
print(f"   Std Dev: {rep_performance['win_rate_pct'].std():.1f}%")

print("\n\n📋 Top 5 Performers:")
print(rep_performance.head()[['sales_rep_id', 'total_deals', 'won_deals', 'win_rate_pct', 'avg_sales_cycle']].to_string(index=False))

print("\n\n📋 Bottom 5 Performers:")
print(rep_performance.tail()[['sales_rep_id', 'total_deals', 'won_deals', 'win_rate_pct', 'avg_sales_cycle']].to_string(index=False))
```

```python
# Rep Performance Visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Win Rate Distribution
ax1 = axes[0]
colors = ['#28A745' if v >= overall_win_rate else '#DC3545' for v in rep_performance['win_rate_pct']]
ax1.bar(range(len(rep_performance)), rep_performance['win_rate_pct'], color=colors, alpha=0.7)
ax1.axhline(y=overall_win_rate, color='navy', linestyle='--', linewidth=2, label=f'Company Avg: {overall_win_rate:.1f}%')
ax1.set_xlabel('Sales Rep (ranked by performance)')
ax1.set_ylabel('Win Rate (%)')
ax1.set_title('📊 Sales Rep Win Rate Distribution', fontsize=12, fontweight='bold')
ax1.legend()

# Win Rate vs. Deal Volume scatter
ax2 = axes[1]
scatter = ax2.scatter(rep_performance['total_deals'], rep_performance['win_rate_pct'], 
                      s=rep_performance['total_revenue']/50000, alpha=0.6, c=rep_performance['win_rate_pct'],
                      cmap='RdYlGn', edgecolors='black')
ax2.axhline(y=overall_win_rate, color='navy', linestyle='--', linewidth=2, alpha=0.7)
ax2.set_xlabel('Total Deals Handled')
ax2.set_ylabel('Win Rate (%)')
ax2.set_title('📊 Win Rate vs. Volume (bubble = revenue)', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

### 2.4 Temporal Deep Dive - Quarter-over-Quarter Analysis

```python
# Quarter-over-Quarter Win Rate by Key Segments
def qoq_analysis(df, segment_col):
    """Analyze win rate changes quarter over quarter by segment"""
    pivot = df.pivot_table(
        values='is_won',
        index=segment_col,
        columns='closed_quarter',
        aggfunc=['mean', 'count']
    )
    
    # Win rates
    wr_df = pivot['mean'] * 100
    
    # Calculate change from first to last quarter
    quarters = wr_df.columns.tolist()
    if len(quarters) >= 2:
        wr_df['Change'] = wr_df[quarters[-1]] - wr_df[quarters[0]]
    
    return wr_df.round(1)

print("=" * 60)
print("QUARTER-OVER-QUARTER WIN RATE BY INDUSTRY")
print("=" * 60)
print(qoq_analysis(df, 'industry'))
```

```python
print("=" * 60)
print("QUARTER-OVER-QUARTER WIN RATE BY REGION")
print("=" * 60)
print(qoq_analysis(df, 'region'))
```

```python
print("=" * 60)
print("QUARTER-OVER-QUARTER WIN RATE BY PRODUCT TYPE")
print("=" * 60)
print(qoq_analysis(df, 'product_type'))
```

```python
print("=" * 60)
print("QUARTER-OVER-QUARTER WIN RATE BY LEAD SOURCE")
print("=" * 60)
print(qoq_analysis(df, 'lead_source'))
```

### 2.5 Sales Cycle Analysis

```python
# Sales Cycle Impact on Outcomes
print("=" * 60)
print("SALES CYCLE ANALYSIS")
print("=" * 60)

won_cycle = df[df['is_won'] == 1]['sales_cycle_days'].describe()
lost_cycle = df[df['is_won'] == 0]['sales_cycle_days'].describe()

print("\n📊 Sales Cycle Comparison (Won vs. Lost Deals):")
comparison = pd.DataFrame({
    'Won Deals': won_cycle,
    'Lost Deals': lost_cycle,
    'Difference': lost_cycle - won_cycle
})
print(comparison.round(1))
```

```python
# Sales Cycle Visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Distribution by outcome
ax1 = axes[0]
ax1.hist([df[df['is_won']==1]['sales_cycle_days'], df[df['is_won']==0]['sales_cycle_days']], 
         bins=30, label=['Won', 'Lost'], alpha=0.7, color=['#28A745', '#DC3545'])
ax1.set_xlabel('Sales Cycle (Days)')
ax1.set_ylabel('Frequency')
ax1.set_title('📊 Sales Cycle Distribution by Outcome', fontsize=12, fontweight='bold')
ax1.legend()

# Win rate by sales cycle bucket
ax2 = axes[1]
df['cycle_bucket'] = pd.cut(df['sales_cycle_days'], 
                            bins=[0, 30, 60, 90, 120, np.inf],
                            labels=['0-30', '31-60', '61-90', '91-120', '120+'])
cycle_wr = df.groupby('cycle_bucket')['is_won'].mean() * 100
colors = ['#28A745' if v >= overall_win_rate else '#DC3545' for v in cycle_wr.values]
bars = ax2.bar(cycle_wr.index.astype(str), cycle_wr.values, color=colors, alpha=0.7)
ax2.axhline(y=overall_win_rate, color='navy', linestyle='--', linewidth=2, label=f'Avg: {overall_win_rate:.1f}%')
ax2.set_xlabel('Sales Cycle (Days)')
ax2.set_ylabel('Win Rate (%)')
ax2.set_title('📊 Win Rate by Sales Cycle Length', fontsize=12, fontweight='bold')
ax2.legend()

plt.tight_layout()
plt.show()
```

### 2.6 Deal Value Analysis

```python
# Deal Value Impact on Outcomes
print("=" * 60)
print("DEAL VALUE ANALYSIS")
print("=" * 60)

# Create deal size buckets
df['deal_size_bucket'] = pd.cut(df['deal_amount'], 
                                 bins=[0, 10000, 25000, 50000, 100000, np.inf],
                                 labels=['<$10K', '$10K-25K', '$25K-50K', '$50K-100K', '$100K+'])

deal_size_wr = df.groupby('deal_size_bucket').agg(
    total_deals=('deal_id', 'count'),
    won_deals=('is_won', 'sum'),
    win_rate=('is_won', 'mean'),
    avg_sales_cycle=('sales_cycle_days', 'mean')
).reset_index()

deal_size_wr['win_rate_pct'] = deal_size_wr['win_rate'] * 100
print(deal_size_wr.to_string(index=False))
```

```python
# Deal Value Visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Win Rate by Deal Size
ax1 = axes[0]
colors = ['#28A745' if v >= overall_win_rate else '#DC3545' for v in deal_size_wr['win_rate_pct']]
bars = ax1.bar(deal_size_wr['deal_size_bucket'].astype(str), deal_size_wr['win_rate_pct'], color=colors, alpha=0.7)
ax1.axhline(y=overall_win_rate, color='navy', linestyle='--', linewidth=2, label=f'Avg: {overall_win_rate:.1f}%')
ax1.set_xlabel('Deal Size')
ax1.set_ylabel('Win Rate (%)')
ax1.set_title('📊 Win Rate by Deal Size', fontsize=12, fontweight='bold')
ax1.legend()

# Volume by Deal Size
ax2 = axes[1]
ax2.bar(deal_size_wr['deal_size_bucket'].astype(str), deal_size_wr['total_deals'], color='#2E86AB', alpha=0.7)
ax2.set_xlabel('Deal Size')
ax2.set_ylabel('Number of Deals')
ax2.set_title('📊 Deal Volume by Size', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

### 2.7 Deal Stage Analysis
